```python
import math
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.animation as animation

from mpl_toolkits.mplot3d import axes3d
```

```python
colors = ['red', 'green', 'blue']
    
fig = plt.figure()
ax = fig.gca(projection = '3d')

lines = []

x = np.array([0, 1])
y = np.array([0, 0])
z = np.array([0, 0])

lines.append([x, y, z])

x = np.array([0, 0])
y = np.array([0, 1])
z = np.array([0, 0])

lines.append([x, y, z])

x = np.array([0, 0])
y = np.array([0, 0])
z = np.array([0, 1])

lines.append([x, y, z])

for i in range(3):
    ax.plot(lines[i][0], lines[i][1], lines[i][2], c = colors[i])

ax.set_xlim3d([0.0, 3.0])
ax.set_ylim3d([0.0, 3.0])
ax.set_zlim3d([0.0, 3.0])

ax.set_title('SLerp interpolacija')

def animation_frame(i):
    lines = []

    x = np.array([0, 1]) + 1
    y = np.array([0, 0]) + 1
    z = np.array([0, 0]) + 1

    lines.append([x, y, z])

    x = np.array([0, 0]) + 1
    y = np.array([0, 1]) + 1
    z = np.array([0, 0]) + 1

    lines.append([x, y, z])

    x = np.array([0, 0]) + 1
    y = np.array([0, 0]) + 1
    z = np.array([0, 1]) + 1

    lines.append([x, y, z])

line_animation = animation.FuncAnimation(fig, animation_frame, 25, interval = 50, blit=False)

plt.show()
```
